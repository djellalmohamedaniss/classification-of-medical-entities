```python
import pandas as pd
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords

sns.set(rc={'figure.figsize':(16,8)})
```

```python
french_stopwords = stopwords.words('french')
```

```python
train_df = pd.read_csv("../datasets/train.csv")
x_train = train_df["RawText"].tolist()
train_labels = [code.lower() for code in train_df["ICD10"].tolist()]
```

```python
def get_vocabulary(data: list[str], separator=' ', processing=False) -> list[str]:
    if processing:
        data = [stem_remove_stopwords(sentence) for sentence in data]
    vocabulary_list = []
    for sentence in data:
        vocabulary_list.extend([token.lower().strip() for token in sentence.split(separator) if len(token.strip()) > 0])
    return list(vocabulary_list)
```

```python
x_vocabulary = get_vocabulary(x_train)
len(x_vocabulary)
```

```python
counter_x_train = Counter(x_vocabulary)
sorted_occurences = {k: v for k, v in sorted(counter_x_train.items(), key=lambda item: item[1], reverse=True)}
```

```python
top_20_words = pd.DataFrame(sorted_occurences.items(), columns=['word', 'occurences'])[:20]
```

```python
ax = sns.barplot(x="word", y="occurences", data=top_20_words)
ax.tick_params(axis='x', rotation=45)
ax.get_figure().savefig('word_occurences.png')
```

```python
wordcloud = WordCloud(background_color = 'white').generate(" ".join(x_train))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig('word_cloud.png')
```

```python
x_vocabulary_no_stopwords = [word for word in x_vocabulary if word not in french_stopwords]
```

```python
counter_x_train = Counter(x_vocabulary_no_stopwords)
sorted_occurences = {k: v for k, v in sorted(counter_x_train.items(), key=lambda item: item[1], reverse=True)}
top_20_words = pd.DataFrame(sorted_occurences.items(), columns=['word', 'occurences'])[:20]
```

```python
ax = sns.barplot(x="word", y="occurences", data=top_20_words)
ax.tick_params(axis='x', rotation=45)
ax.get_figure().savefig('word_occurences_no_stopwords.png')
```
